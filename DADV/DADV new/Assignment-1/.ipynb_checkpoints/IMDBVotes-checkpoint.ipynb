# 1.find the 20 most popular movies with a rank more than 8.0

```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv("E:/DownloadedData/Ratings.tsv", delimiter = '\t')
```

```python
ratings.head()
```

```python
basics = pd.read_csv("E:/DownloadedData/Basics.tsv", delimiter = '\t')
```

```python
basics.head()
```

```python
topRatedMovies = ratings[ratings['averageRating'] > 8.0]
topRatedMovies
```

```python
SortedTopMovies20 = topRatedMovies.sort_values('numVotes', ascending = False).head(20)
SortedTopMovies20
```

```python
ratingMergedBasics = pd.merge(SortedTopMovies20, basics, on= 'tconst')
ratingMergedBasics
```

```python
onlyMovies = ratingMergedBasics.loc[ratingMergedBasics['titleType'] == 'movie']
```

```python
top20Movies = onlyMovies[['originalTitle','averageRating' ]].head(20)
top20Movies
```

```python
top20MoviesSorted =top20Movies.sort_values('averageRating', ascending = False)
top20MoviesSorted
```

```python
top20MoviesSorted.to_csv('D:/DataScience/DADV/DADV new/Assignment-1/top20Movies.csv' )
```

# 2.	find the 20 best rated movies with over 40,000 votes in the 2000s (year >= 2000)

```python
basics.dtypes
```

```python
basics['startYear'] = pd.to_numeric(basics['startYear'], errors='coerce')
```

```python
basics.dtypes
```

```python
basicsAfter2000 = basics[basics.startYear >= 2000]
basicsAfter2000
```

```python
basicsafter2000MergedRatings = pd.merge(basicsAfter2000, ratings, on= 'tconst')
basicsafter2000MergedRatings
```

```python
basicsMin40000votes = basicsafter2000MergedRatings[basicsafter2000MergedRatings['numVotes'] > 40000]
basicsMin40000votes
```

```python
basicsMin40000votesSorted = basicsMin40000votes.sort_values('averageRating', ascending = False)
basicsMin40000votesSorted
```

```python
filterMovies = basicsMin40000votesSorted.loc[basicsMin40000votesSorted['titleType'] == 'movie']
filterMovies
```

```python
top20Movies = filterMovies[['originalTitle','numVotes', 'startYear','averageRating' ]].head(20)
top20Movies
```

```python
top20Movies.to_csv('E:/Spec/Data-Science/DADV/DADV new/Assignment-1/MovieWithMostVotes.csv')
```
