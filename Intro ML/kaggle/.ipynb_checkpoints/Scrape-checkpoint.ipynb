```python
from pandas import *
import requests
import bs4
```

```python
url='https://finance.yahoo.com/quote/AAPL/history?period1=1519842600&period2=1544639400&interval=1d&filter=history&frequency=1d&guccounter=1'
source=requests.get(url).text
info=bs4.BeautifulSoup(source)
url=[]

tc=info.find_all('table',{'class': 'W(100%)'})
# print(tc[0])
p=tc[0].find_all('span')
s = ""
s = p
for i in s:
    q = str(i).split('>')
    z = q[1].split("<")
    z = z[0].split(",")
    url.append(z[0])
    
url

```
